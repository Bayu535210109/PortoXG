```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import os
from xgboost import XGBRegressor, plot_importance
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from ta.momentum import RSIIndicator, ROCIndicator
from ta.trend import MACD, EMAIndicator, WMAIndicator
from ta.volatility import BollingerBands
from datetime import datetime, timedelta
from tqdm import tqdm


# Daftar 15 saham yang akan diprediksi
tickers = ['NVDA', 'AAPL', 'GOOGL', '005930.KS', '9988.HK', '000660.KQ',
          'AMD', '0700.HK', 'TSLA', 'ORCL', 'AMZN', 'INTC', 'MSFT']
sp500_ticker = '^GSPC'
prediksi_global = pd.DataFrame()
start_date = '2023-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

# Membuat folder untuk menyimpan model
if not os.path.exists('models'):
    os.makedirs('models')

# Variabel untuk menyimpan model dan scaler dari ticker terakhir saja
last_model = None
last_scaler = None

# Fungsi untuk memproses saham
def process_stock(ticker):
    print(f"\n{'='*20} MEMPROSES {ticker} {'='*20}")
    
    features_path = "models/features_global.pkl"
    
    try:
        # === AMBIL DATA SAHAM & S&P 500 ===
        df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
        sp500 = yf.download(sp500_ticker, start=start_date, end=end_date, auto_adjust=False)
        
        if df.empty or len(df) < 100:
            print(f"Data tidak cukup untuk {ticker}. Melewati...")
            return {
                "ticker": ticker,
                "status": "Gagal - Data tidak cukup",
                "last_close": None,
                "prediction": None,
                "future_predictions": None,
                "mae": None, 
                "rmse": None, 
                "r2": None
            }
        
        # === PRA-PROSES ===
        df = df[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()
        sp500 = sp500[['Close']].rename(columns={'Close': 'SP500_Close'}).dropna()
        df = df.join(sp500, how='inner')  # gabungkan dengan SP500
        
        # === FITUR TEKNIKAL & INTERAKSI ===
        df['Volume'] = np.log1p(df['Volume'])
        df['Lag1'] = df['Close'].shift(1)
        df['Lag2'] = df['Close'].shift(2)
        df['Lag3'] = df['Close'].shift(3)
        df['Lag4'] = df['Close'].shift(4)
        df['MA5'] = df['Close'].rolling(5).mean()
        df['MA10'] = df['Close'].rolling(10).mean()
        df['MA20'] = df['Close'].rolling(20).mean()
        df['MA50'] = df['Close'].rolling(50).mean()
        df['EMA10'] = EMAIndicator(close=df['Close'].squeeze(), window=10).ema_indicator()
        df['EMA20'] = EMAIndicator(close=df['Close'].squeeze(), window=20).ema_indicator()
        df['EMA50'] = EMAIndicator(close=df['Close'].squeeze(), window=50).ema_indicator()
        df['WMA10'] = WMAIndicator(close=df['Close'].squeeze(), window=10).wma()
        df['WMA20'] = WMAIndicator(close=df['Close'].squeeze(), window=20).wma()
        df['Momentum'] = df['Close'] - df['Close'].shift(4)
        df['ROC'] = ROCIndicator(close=df['Close'].squeeze(), window=5).roc()
        df['Volatility'] = df['Close'].pct_change().rolling(window=5).std()
        df['High_Low'] = df['High'] - df['Low']
        df['Close_Open'] = df['Close'] - df['Open']
        df['Rolling_Max_5'] = df['Close'].rolling(5).max()
        df['Rolling_Min_5'] = df['Close'].rolling(5).min()
        df['Price_Change'] = df['Close'].diff()
        df['3D_Trend'] = df['Close'] - df['Close'].shift(3)
        df['Lag1_ratio'] = df['Lag1'] / df['Lag2']
        df['Lag2_ratio'] = df['Lag3'] / df['Lag4']
        df['MA5_vs_EMA10'] = df['MA5'] - df['EMA10']
        df['MA5_to_MA20'] = df['MA5'] / df['MA20']
        df['EMA10_to_EMA50'] = df['EMA10'] / df['EMA50']
        df['WMA10_to_WMA20'] = df['WMA10'] / df['WMA20']
        df['RSI'] = RSIIndicator(close=df['Close'].squeeze(), window=14).rsi()
        macd = MACD(close=df['Close'].squeeze())
        df['MACD'] = macd.macd()
        df['MACD_signal'] = macd.macd_signal()
        bb = BollingerBands(close=df['Close'].squeeze())
        df['BB_high'] = bb.bollinger_hband()
        df['BB_low'] = bb.bollinger_lband()
        df['Year'] = df.index.year
        df['Month'] = df.index.month
        df['Day'] = df.index.day
        df['DayOfWeek'] = df.index.dayofweek
        df['Return'] = df['Close'].pct_change()
        
        # === TARGET: Harga Penutupan Esok Hari ===
        df['Target'] = df['Close'].shift(-1)
        df.dropna(inplace=True)
        
        # === FITUR MODEL ===
        features = [col for col in df.columns if col not in ['Target', 'Close']]
        X = df[features]
        y = df['Target']

        # ⬇ Tambahkan di sini:
        if not os.path.exists(features_path):
            clean_features = [col if isinstance(col, str) else col[0] for col in df.columns]
            clean_features = [f for f in clean_features if f not in ['Close', 'Target']]
            joblib.dump(list(clean_features), features_path)
            print(f"✅ Fitur global disimpan: {features_path}")


        # === TRAIN-TEST SPLIT TANPA SHUFFLE (80:20) ===
        split_idx = int(len(df) * 0.8)
        X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
        y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

        # === NORMALISASI ===
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # === TRAINING DENGAN TIME SERIES SPLIT PADA DATA TRAIN SAJA ===
        tscv = TimeSeriesSplit(n_splits=3)
        param_grid = {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [5, 7],
            'subsample': [0.8],
            'colsample_bytree': [0.8]
        }
        model = XGBRegressor(random_state=42)
        grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=tscv, verbose=0)
        grid_search.fit(X_train_scaled, y_train)
        best_model = grid_search.best_estimator_

        # === EVALUASI DI TEST SET ===
        y_pred_test = best_model.predict(X_test_scaled)
        mae = mean_absolute_error(y_test, y_pred_test)
        mse = mean_squared_error(y_test, y_pred_test)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred_test)
        mape_test = mean_absolute_percentage_error(y_test, y_pred_test) * 100

        
        # === PREDIKSI HARI BERIKUTNYA (skip weekend) ===
        last_row = X_test_scaled[-1].reshape(1, -1)
        last_close_val = df['Close'].iloc[-1].item()
        predicted_close = best_model.predict(last_row)[0].item()
        final_pred = predicted_close
        last_date = df.index[-1]
        next_date = last_date + timedelta(days=1)
        while next_date.weekday() >= 5:
            next_date += timedelta(days=1)
        
        # === CETAK HASIL ===
        print(f"\nMAE: {mae:.4f}")
        print(f"MSE: {mse:.4f}")
        print(f"RMSE: {rmse:.4f}")
        print(f"R² : {r2:.4f}")
        print(f"Tanggal penutupan terakhir: {last_date.date()}")
        print(f"Harga penutupan terakhir  : {last_close_val:.2f}")
        print(f"Prediksi penutupan: {final_pred:.2f} (untuk tanggal {next_date.date()})")
        
        # === CEK HARGA AKTUAL BESOK (JIKA SUDAH ADA) ===
        actual_data = yf.download(ticker, start=next_date.strftime('%Y-%m-%d'), end=(next_date + timedelta(days=1)).strftime('%Y-%m-%d'))
        
        harga_aktual_besok = None
        selisih = None
        mape = None
        akurasi = None
        
        if not actual_data.empty:
            harga_aktual_besok = actual_data['Close'].iloc[0].item()
            selisih = abs(final_pred - harga_aktual_besok)
            mape = (selisih / harga_aktual_besok) * 100 if harga_aktual_besok != 0 else 0
            akurasi = 100 - mape
        
            print(f"\nEvaluasi Real (Auto Yahoo Finance):")
            print(f"Harga aktual              : {harga_aktual_besok:.2f}")
            print(f"Selisih prediksi          : {selisih:.2f}")
            print(f"MAPE terhadap aktual      : {mape:.2f}%")
            print(f"Akurasi prediksi          : {akurasi:.2f}%")
        else:
            print(f"\nHarga aktual untuk tanggal {next_date.date()} belum tersedia di Yahoo Finance.")

        # Simpan model & scaler terakhir
        global last_model, last_scaler
        last_model = best_model
        last_scaler = scaler

        # === PREDIKSI 3 BULAN KE DEPAN (tanpa tanggal duplikat & skip weekend) ===
        future_predictions = []
        future_dates = []
        
        X_latest = X.iloc[-1:].copy()
        future_date = df.index[-1]  # mulai dari tanggal terakhir
        
        for _ in range(90):
            # Geser ke hari berikutnya dan lewati weekend
            future_date += timedelta(days=1)
            while future_date.weekday() >= 5:  # 5 = Sabtu, 6 = Minggu
                future_date += timedelta(days=1)
        
            # Prediksi harga
            X_scaled_latest = scaler.transform(X_latest)
            pred = best_model.predict(X_scaled_latest)[0]
        
            # Gabung prediksi dengan MA5 (ensemble)
            ma5_latest = X_latest['MA5'].values[0]
            final_pred = 0.7 * pred + 0.3 * ma5_latest
        
            # Simpan hasil
            future_predictions.append(final_pred)
            future_dates.append(future_date.strftime('%Y-%m-%d'))
        
            # Update fitur untuk hari selanjutnya
            new_row = X_latest.copy()
            new_row['Lag4'] = new_row['Lag3']
            new_row['Lag3'] = new_row['Lag2']
            new_row['Lag2'] = new_row['Lag1']
            new_row['Lag1'] = final_pred
            new_row['MA5'] = (new_row['MA5'] * 4 + final_pred) / 5
            new_row['MA10'] = (new_row['MA10'] * 9 + final_pred) / 10
            new_row['MA20'] = (new_row['MA20'] * 19 + final_pred) / 20
            new_row['MA50'] = (new_row['MA50'] * 49 + final_pred) / 50
            new_row['Momentum'] = final_pred - new_row['Lag4']
            new_row['Lag1_ratio'] = new_row['Lag1'] / new_row['Lag2']
            new_row['Lag2_ratio'] = new_row['Lag3'] / new_row['Lag4']
            new_row['MA5_vs_EMA10'] = new_row['MA5'] - new_row['EMA10']
            new_row['MA5_to_MA20'] = new_row['MA5'] / new_row['MA20']
            new_row['EMA10_to_EMA50'] = new_row['EMA10'] / new_row['EMA50']
            new_row['WMA10_to_WMA20'] = new_row['WMA10'] / new_row['WMA20']
            new_row['3D_Trend'] = final_pred - new_row['Lag3']
            new_row['Price_Change'] = final_pred - new_row['Lag1']
            new_row['Return'] = new_row['Price_Change'] / new_row['Lag1']
        
            # Update tanggal
            new_row['Day'] = future_date.day
            new_row['Month'] = future_date.month
            new_row['Year'] = future_date.year
            new_row['DayOfWeek'] = future_date.weekday()
        
            # Ganti untuk iterasi selanjutnya
            X_latest = new_row
        
        # === CETAK HASIL PREDIKSI ===
        print("\nPrediksi 3 Bulan ke depan:")
        for tanggal, harga in zip(future_dates, future_predictions):
            print(f"Prediksi untuk tanggal {tanggal} : {harga:.2f}")
        
        # --- Tambahkan ke DataFrame global ---
        df_pred_ticker = pd.DataFrame({
            'Tanggal Prediksi': future_dates,
            ticker: [round(p, 2) for p in future_predictions]
        })

        global prediksi_global
        if prediksi_global.empty:
            prediksi_global = df_pred_ticker
        else:
            prediksi_global = pd.merge(prediksi_global, df_pred_ticker, on='Tanggal Prediksi', how='outer')

        # === EVALUASI HISTORIS TERHADAP 90 Data TERAKHIR (VALIDASI MODEL) ===
        X_val = X_test_scaled[-90:]
        y_val = y_test[-90:]
        y_pred_val = best_model.predict(X_val)
        
        mae_val = mean_absolute_error(y_val, y_pred_val)
        mse_val = mean_squared_error(y_val, y_pred_val)
        rmse_val = np.sqrt(mse_val)
        r2_val = r2_score(y_val, y_pred_val)
        
        print("\n=== Evaluasi Model Terhadap 3 Bulan Terakhir (Validasi Historis) ===")
        print(f"MAE   : {mae_val:.4f}")
        print(f"MSE   : {mse_val:.4f}")
        print(f"RMSE  : {rmse_val:.4f}")
        print(f"R²    : {r2_val:.4f}")
        
        # --- Ambil 30 hari terakhir dari data historis aktual ---
        last_n = 30
        df_plot = df[-last_n:].copy()
        df_plot_dates = df_plot.index.strftime('%Y-%m-%d').tolist()
        
        # --- Konversi prediksi ke float secara aman ---
        future_predictions_float = []
        
        for i, p in enumerate(future_predictions):
            try:
                # Flatten jika array atau list
                if isinstance(p, (list, np.ndarray)):
                    p = np.ravel(p)[0]
                future_predictions_float.append(float(p))
            except Exception as e:
                print(f"❗ Error konversi prediksi ke-{i}: {p} -> {e}")
        
        # --- Gabungkan tanggal ---
        extended_dates = df_plot_dates + future_dates
        
        # --- Gabungkan nilai aktual + prediksi dan pastikan semuanya float ---
        combined_raw = df_plot['Close'].values.tolist() + future_predictions_float
        extended_values = []
        
        for i, v in enumerate(combined_raw):
            try:
                if isinstance(v, (list, np.ndarray)):
                    v = np.ravel(v)[0]
                extended_values.append(float(v))
            except Exception as e:
                print(f"❗ Gagal convert nilai ke-{i}: {v} -> {e}")
        
        # --- Validasi panjang sama ---
        assert len(extended_dates) == len(extended_values), "Jumlah tanggal dan nilai tidak sama!"
        
        # --- Tentukan batas pembatas aktual vs prediksi ---
        split_index = len(df_plot_dates) - 1
        
        # Tampilkan grafik
        plt.figure(figsize=(12, 5))
        plt.plot(extended_dates, extended_values, label='Harga', marker='o', color='blue')
        
        # --- Garis vertikal pemisah ---
        plt.axvline(x=extended_dates[split_index], color='red', linestyle='--', label='Mulai Prediksi')
        
        # --- Area bayangan untuk aktual dan prediksi ---
        plt.fill_between(extended_dates[:split_index+1],
                        min(extended_values), max(extended_values),
                        color='gray', alpha=0.1, label='Aktual')
        plt.fill_between(extended_dates[split_index:],
                        min(extended_values), max(extended_values),
                        color='orange', alpha=0.1, label='Prediksi')
        
        # --- Format akhir ---
        plt.title(f'Harga Penutupan {ticker}: Aktual vs Prediksi 3 Bulan ke Depan')
        plt.xlabel('Tanggal')
        plt.ylabel('Harga Penutupan')
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()  # Tampilkan grafik langsung daripada menyimpannya
        
        # --- Plot Split Training/Testing ---
        # Tentukan proporsi split (80% training, 20% testing)
        train_size = 0.8
        split_index = int(len(df) * train_size)
        
        # Bagi data
        train_data = df.iloc[:split_index]
        test_data = df.iloc[split_index:]
        
        # Plot Data
        plt.figure(figsize=(12, 5))
        plt.plot(train_data.index, train_data['Close'], label='Training Data', color='blue')
        plt.plot(test_data.index, test_data['Close'], label='Testing Data', color='orange')
        
        # Garis vertikal sebagai pemisah
        plt.axvline(x=df.index[split_index], color='red', linestyle='--', label='Split Training/Testing')
        
        # Format akhir plot
        plt.title(f'Harga Penutupan {ticker}: Data Training vs Data Testing')
        plt.xlabel('Tanggal')
        plt.ylabel('Harga Penutupan')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.xticks(rotation=45)
        plt.show()  # Tampilkan grafik langsung
        
        # Kembalikan ringkasan hasil
        return {
            "ticker": ticker,
            "status": "Sukses",
            "last_date": last_date.strftime('%Y-%m-%d'),
            "last_close": round(last_close_val, 2),
            "next_date": next_date.strftime('%Y-%m-%d'),
            "prediction": round(final_pred, 2),
            "future_dates": future_dates,
            "future_predictions": [round(p, 2) for p in future_predictions],
            "actual_price": round(harga_aktual_besok, 2) if harga_aktual_besok is not None else None,
            "selisih": round(selisih, 2) if selisih is not None else None,
            "akurasi": round(akurasi, 2) if akurasi is not None else None,
            "mae": round(mae, 4), 
            "rmse": round(rmse, 4),
            "r2": round(r2, 4)
        }
        
    except Exception as e:
        print(f"❌ Error saat memproses {ticker}: {str(e)}")
        return {
            "ticker": ticker,
            "status": f"Gagal - {str(e)}",
            "last_close": None,
            "prediction": None
        }

# Jalankan proses untuk semua saham
all_results = []
for ticker in tqdm(tickers, desc="Memproses saham"):
    result = process_stock(ticker)
    all_results.append(result)

# Simpan hanya model terakhir sebagai model global
if last_model is not None and last_scaler is not None:
    joblib.dump(last_model, "models/model_xgb_global_13saham.pkl")
    joblib.dump(last_scaler, "models/scaler_global_13saham.pkl")
    print("\n✅ Model global 13 saham berhasil disimpan (diambil dari proses terakhir).")

# Simpan hasil prediksi gabungan
prediksi_global = prediksi_global[['Tanggal Prediksi'] + tickers]
prediksi_global.to_csv(f"prediksi_3bulan_13saham.csv", index=False)
print(f"\n✅ File prediksi gabungan disimpan ke: prediksi_3bulan_13saham.csv")

# Buat laporan ringkasan
print("\n" + "="*70)
print(" "*20 + "RINGKASAN PREDIKSI SAHAM" + " "*20)
print("="*70)
print(f"{'Ticker':<7} {'Harga Terakhir':<15} {'Prediksi':<15} {'Selisih':<10} {'Akurasi':<10} {'MAE':<10} {'R²':<10}")
print("-"*70)

for result in all_results:
    if result["status"].startswith("Sukses"):
        print(f"{result['ticker']:<7} {result['last_close']:<15} {result['prediction']:<15} {result['selisih'] if result['selisih'] is not None else 'N/A':<10} {result['akurasi'] if result['akurasi'] is not None else 'N/A':<10} {result['mae']:<10} {result['r2']:<10}")
    else:
        print(f"{result['ticker']:<7} {'GAGAL':<15} {result['status']}")
print("="*70)

# Visualisasi performa model untuk semua saham
valid_results = [r for r in all_results if r["status"].startswith("Sukses")]
if valid_results:
    tickers = [r['ticker'] for r in valid_results]
    mae_values = [r['mae'] for r in valid_results]
    r2_values = [r['r2'] for r in valid_results]
    
    # Plot performa model (MAE & R²)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot MAE
    ax1.bar(tickers, mae_values, color='skyblue')
    ax1.set_title('MAE untuk Setiap Saham')
    ax1.set_xlabel('Saham')
    ax1.set_ylabel('MAE (Mean Absolute Error)')
    ax1.set_xticklabels(tickers, rotation=45)
    
    # Plot R²
    ax2.bar(tickers, r2_values, color='lightgreen')
    ax2.set_title('R² untuk Setiap Saham')
    ax2.set_xlabel('Saham')
    ax2.set_ylabel('R² Score')
    ax2.set_xticklabels(tickers, rotation=45)
    
    plt.tight_layout()
    plt.show()  # Tampilkan grafik performa secara langsung
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from datetime import datetime, timedelta
from ta.momentum import RSIIndicator, ROCIndicator
from ta.trend import MACD, EMAIndicator, WMAIndicator
from ta.volatility import BollingerBands

# Load model dan scaler yang sudah disimpan
model = joblib.load("models/model_xgb_global_13saham.pkl")
scaler = joblib.load("models/scaler_global_13saham.pkl")
features = joblib.load("models/features_global.pkl")

# Cetak fitur yang digunakan oleh model
print("Fitur yang digunakan oleh model:")
print(features)

# Fungsi untuk melakukan prediksi
def prediksi_saham(ticker, hari_prediksi=30):
    print(f"\n{'='*20} MEMPREDIKSI {ticker} {'='*20}")
    
    # Ambil data historis (satu tahun terakhir untuk fitur teknikal)
    start_date = (datetime.today() - timedelta(days=365)).strftime('%Y-%m-%d')
    end_date = datetime.today().strftime('%Y-%m-%d')
    
    try:
        # Download data saham dan S&P 500
        df = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)
        sp500 = yf.download('^GSPC', start=start_date, end=end_date, auto_adjust=False)
        
        if df.empty or len(df) < 100:
            print(f"Data tidak cukup untuk {ticker}. Minimal butuh 100 hari data.")
            return None
        
        # Pra-proses data
        df = df[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()
        sp500 = sp500[['Close']].rename(columns={'Close': 'SP500_Close'}).dropna()
        df = df.join(sp500, how='inner')  # gabungkan dengan SP500
        
        # Membuat fitur teknikal yang sama dengan model pelatihan
        df['Volume'] = np.log1p(df['Volume'])
        df['Lag1'] = df['Close'].shift(1)
        df['Lag2'] = df['Close'].shift(2)
        df['Lag3'] = df['Close'].shift(3)
        df['Lag4'] = df['Close'].shift(4)
        df['MA5'] = df['Close'].rolling(5).mean()
        df['MA10'] = df['Close'].rolling(10).mean()
        df['MA20'] = df['Close'].rolling(20).mean()
        df['MA50'] = df['Close'].rolling(50).mean()
        df['EMA10'] = EMAIndicator(close=df['Close'].squeeze(), window=10).ema_indicator()
        df['EMA20'] = EMAIndicator(close=df['Close'].squeeze(), window=20).ema_indicator()
        df['EMA50'] = EMAIndicator(close=df['Close'].squeeze(), window=50).ema_indicator()
        df['WMA10'] = WMAIndicator(close=df['Close'].squeeze(), window=10).wma()
        df['WMA20'] = WMAIndicator(close=df['Close'].squeeze(), window=20).wma()
        df['Momentum'] = df['Close'] - df['Close'].shift(4)
        df['ROC'] = ROCIndicator(close=df['Close'].squeeze(), window=5).roc()
        df['Volatility'] = df['Close'].pct_change().rolling(window=5).std()
        df['High_Low'] = df['High'] - df['Low']
        df['Close_Open'] = df['Close'] - df['Open']
        df['Rolling_Max_5'] = df['Close'].rolling(5).max()
        df['Rolling_Min_5'] = df['Close'].rolling(5).min()
        df['Price_Change'] = df['Close'].diff()
        df['3D_Trend'] = df['Close'] - df['Close'].shift(3)
        df['Lag1_ratio'] = df['Lag1'] / df['Lag2']
        df['Lag2_ratio'] = df['Lag3'] / df['Lag4']
        df['MA5_vs_EMA10'] = df['MA5'] - df['EMA10']
        df['MA5_to_MA20'] = df['MA5'] / df['MA20']
        df['EMA10_to_EMA50'] = df['EMA10'] / df['EMA50']
        df['WMA10_to_WMA20'] = df['WMA10'] / df['WMA20']
        df['RSI'] = RSIIndicator(close=df['Close'].squeeze(), window=14).rsi()
        macd = MACD(close=df['Close'].squeeze())
        df['MACD'] = macd.macd()
        df['MACD_signal'] = macd.macd_signal()
        bb = BollingerBands(close=df['Close'].squeeze())
        df['BB_high'] = bb.bollinger_hband()
        df['BB_low'] = bb.bollinger_lband()
        df['Year'] = df.index.year
        df['Month'] = df.index.month
        df['DayOfWeek'] = df.index.dayofweek
        # Add the missing features
        df['Day'] = df.index.day  # Add day of month
        df['Return'] = df['Close'].pct_change()  # Add daily return
        df['Close_copy'] = df['Close'].squeeze()  # Add a copy of Close as a feature
        df['Close_MA5_ratio'] = df['Close'].squeeze() / df['MA5'].squeeze() # Add another potential missing feature

        
        # Hapus data NaN
        df.dropna(inplace=True)
        
        # Ambil data terakhir untuk fitur prediksi
        latest_data = df.iloc[-1:]
        
        # Pastikan kita hanya menggunakan fitur yang ada di features list
        available_features = []
        for feature in features:
            if feature in df.columns:
                available_features.append(feature)
            else:
                print(f"Fitur '{feature}' tidak ditemukan dalam data. Melewatkan...")
        
        X_latest = latest_data[available_features]
        
        # Normalisasi data
        X_scaled = scaler.transform(X_latest)
        
        # Prediksi untuk hari berikutnya
        last_close = df['Close'].iloc[-1]
        next_date = df.index[-1] + timedelta(days=1)
        while next_date.weekday() >= 5:  # Skip weekend
            next_date += timedelta(days=1)
            
        prediction = model.predict(X_scaled)[0]
        
        print(f"Tanggal terakhir data: {df.index[-1].strftime('%Y-%m-%d')}")
        print(f"Harga penutupan terakhir: {last_close:.2f}")
        print(f"Prediksi untuk {next_date.strftime('%Y-%m-%d')}: {prediction:.2f}")
        
        # Prediksi untuk beberapa hari ke depan
        future_predictions = []
        future_dates = []
        
        X_pred = X_latest.copy()
        future_date = df.index[-1]
        
        for _ in range(hari_prediksi):
            # Geser ke hari berikutnya dan lewati weekend
            future_date += timedelta(days=1)
            while future_date.weekday() >= 5:  # 5 = Sabtu, 6 = Minggu
                future_date += timedelta(days=1)
            
            # Prediksi harga
            X_scaled_pred = scaler.transform(X_pred)
            pred = model.predict(X_scaled_pred)[0]
            
            # Gabung prediksi dengan MA5 (ensemble)
            ma5_pred = X_pred['MA5'].values[0]
            final_pred = 0.7 * pred + 0.3 * ma5_pred
            
            # Simpan hasil
            future_predictions.append(final_pred)
            future_dates.append(future_date.strftime('%Y-%m-%d'))
            
            # Update fitur untuk hari selanjutnya (hanya fitur yang ada di dataframe)
            if 'Lag4' in X_pred.columns and 'Lag3' in X_pred.columns:
                X_pred['Lag4'] = X_pred['Lag3']
            if 'Lag3' in X_pred.columns and 'Lag2' in X_pred.columns:
                X_pred['Lag3'] = X_pred['Lag2']
            if 'Lag2' in X_pred.columns and 'Lag1' in X_pred.columns:
                X_pred['Lag2'] = X_pred['Lag1']
            if 'Lag1' in X_pred.columns:
                X_pred['Lag1'] = final_pred
            if 'MA5' in X_pred.columns:
                X_pred['MA5'] = (X_pred['MA5'] * 4 + final_pred) / 5
            if 'MA10' in X_pred.columns:
                X_pred['MA10'] = (X_pred['MA10'] * 9 + final_pred) / 10
            if 'MA20' in X_pred.columns:
                X_pred['MA20'] = (X_pred['MA20'] * 19 + final_pred) / 20
            if 'MA50' in X_pred.columns:
                X_pred['MA50'] = (X_pred['MA50'] * 49 + final_pred) / 50
            if 'Momentum' in X_pred.columns and 'Lag4' in X_pred.columns:
                X_pred['Momentum'] = final_pred - X_pred['Lag4']
            if 'Lag1_ratio' in X_pred.columns and 'Lag1' in X_pred.columns and 'Lag2' in X_pred.columns:
                X_pred['Lag1_ratio'] = X_pred['Lag1'] / X_pred['Lag2']
            if 'Lag2_ratio' in X_pred.columns and 'Lag3' in X_pred.columns and 'Lag4' in X_pred.columns:
                X_pred['Lag2_ratio'] = X_pred['Lag3'] / X_pred['Lag4']
            if 'MA5_vs_EMA10' in X_pred.columns and 'MA5' in X_pred.columns and 'EMA10' in X_pred.columns:
                X_pred['MA5_vs_EMA10'] = X_pred['MA5'] - X_pred['EMA10']
            if 'MA5_to_MA20' in X_pred.columns and 'MA5' in X_pred.columns and 'MA20' in X_pred.columns:
                X_pred['MA5_to_MA20'] = X_pred['MA5'] / X_pred['MA20']
            if 'EMA10_to_EMA50' in X_pred.columns and 'EMA10' in X_pred.columns and 'EMA50' in X_pred.columns:
                X_pred['EMA10_to_EMA50'] = X_pred['EMA10'] / X_pred['EMA50']
            if 'WMA10_to_WMA20' in X_pred.columns and 'WMA10' in X_pred.columns and 'WMA20' in X_pred.columns:
                X_pred['WMA10_to_WMA20'] = X_pred['WMA10'] / X_pred['WMA20']
            if '3D_Trend' in X_pred.columns and 'Lag3' in X_pred.columns:
                X_pred['3D_Trend'] = final_pred - X_pred['Lag3']
            if 'Price_Change' in X_pred.columns and 'Lag1' in X_pred.columns:
                X_pred['Price_Change'] = final_pred - X_pred['Lag1']
            
            # Update tanggal
            if 'Month' in X_pred.columns:
                X_pred['Month'] = future_date.month
            if 'Year' in X_pred.columns:
                X_pred['Year'] = future_date.year
            if 'Day' in X_pred.columns:
                X_pred['Day'] = future_date.day
            if 'DayOfWeek' in X_pred.columns:
                X_pred['DayOfWeek'] = future_date.weekday()
            
        # Visualisasi hasil prediksi
        # Ambil 30 hari terakhir dari data historis
        last_n = 30
        df_plot = df[-last_n:].copy()
        df_plot_dates = df_plot.index.strftime('%Y-%m-%d').tolist()
        
        # Gabungkan tanggal dan nilai untuk plot
        extended_dates = df_plot_dates + future_dates
        extended_values = df_plot['Close'].values.tolist() + future_predictions
        
        # Batas pemisah antara data aktual dan prediksi
        split_index = len(df_plot_dates) - 1
        
        # Plot hasil
        plt.figure(figsize=(12, 5))
        plt.plot(extended_dates, extended_values, marker='o', label='Harga', color='blue')
        
        # Garis vertikal pemisah
        plt.axvline(x=extended_dates[split_index], color='red', linestyle='--', label='Mulai Prediksi')
        
        # Area bayangan untuk aktual dan prediksi
        plt.fill_between(extended_dates[:split_index+1],
                        min(extended_values), max(extended_values),
                        color='gray', alpha=0.1, label='Aktual')
        plt.fill_between(extended_dates[split_index:],
                        min(extended_values), max(extended_values),
                        color='orange', alpha=0.1, label='Prediksi')
        
        plt.title(f'Harga Penutupan {ticker}: Aktual vs Prediksi {hari_prediksi} Hari ke Depan')
        plt.xlabel('Tanggal')
        plt.ylabel('Harga Penutupan')
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        
        # Kembalikan hasil prediksi
        return {
            "ticker": ticker,
            "last_date": df.index[-1].strftime('%Y-%m-%d'),
            "last_close": round(last_close, 2),
            "next_date": next_date.strftime('%Y-%m-%d'),
            "next_prediction": round(prediction, 2),
            "future_dates": future_dates,
            "future_predictions": [round(p, 2) for p in future_predictions]
        }
        
    except Exception as e:
        print(f"Error saat memprediksi {ticker}: {str(e)}")
        import traceback
        traceback.print_exc()  # Tampilkan stack trace lengkap
        return None

# Contoh penggunaan
if __name__ == "__main__":
    # Masukkan ticker saham yang ingin diprediksi
    ticker = input("Masukkan kode saham (contoh: AAPL, GOOGL, MSFT): ")
    
    # Masukkan jumlah hari prediksi
    try:
        hari = int(input("Masukkan jumlah hari prediksi (default: 30): ") or "30")
    except ValueError:
        hari = 30
        print("Input tidak valid, menggunakan default 30 hari.")
    
    hasil = prediksi_saham(ticker, hari)
    
    if hasil:
        print("\nRingkasan Prediksi:")
        print(f"Ticker: {hasil['ticker']}")
        print(f"Harga terakhir ({hasil['last_date']}): {hasil['last_close']}")
        print(f"Prediksi berikutnya ({hasil['next_date']}): {hasil['next_prediction']}")
        
        print("\nPrediksi untuk beberapa hari ke depan:")
        for tanggal, harga in zip(hasil['future_dates'], hasil['future_predictions']):
            print(f"{tanggal}: {harga}")
```
